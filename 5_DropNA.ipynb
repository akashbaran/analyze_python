```python
# forward and backward data filling for NA
# getting the NaN values again

import pandas as pd
import datetime
from pandas_datareader import data, wb
start = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2013, 1, 1)

att = data.DataReader("T", "yahoo", start, end)
att.head()

att['50MA'] = att['Close'].rolling(50).mean()
att['10MA'] = att['Close'].rolling(10).mean()
att.head()
```

```python
# hear we can backfill but we will also try forware fill.
att.fillna(method='ffill', inplace=True)
att.head()
```

```python
# no difference above.
att.fillna(method='bfill', inplace=True)
att.head()
```

```python
# back fill works.. now we create this data again and see few other methods.
import pandas as pd
import datetime
from pandas_datareader import data, wb
start = datetime.datetime(2012, 1, 1)
end = datetime.datetime(2013, 1, 1)

att = data.DataReader("T", "yahoo", start, end)
att.head()

att['50MA'] = att['Close'].rolling(50).mean()
att['10MA'] = att['Close'].rolling(10).mean()
att.head()
```

```python
att2 = att.fillna(value=-9999)
att2.head()
# we get all NaN replaced by -9999...
```

```python
# we can add limit to no of rows to be changed. Suppose we want to change first 2 rows.
att3 = att.fillna(value=-9999, limit=2)
att3.head()
```
