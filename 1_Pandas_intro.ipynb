```python
import pandas as pd
```

```python
# keys will column, value will be rows
starting = {
    'Col_1':[1,2,3,4,45,67,43],
    'Col_2':[3,76,5,3,23,54,98],
    'Col_3':[7,98,54,34,1,2,3],
    'Col_4':[1,5,87,0,7,6,5],
    'Col_5':[98,77,76,78,45,55,56],
    'Col_6':[12,13,14,15,16,12,15]
}
# dataframe
df = pd.DataFrame(starting)
print(df)
```

```python
print(df.dtypes)
```

```python
print(df['Col_1'])
```

```python
print(df['Col_1'][3])
```

```python
print(df.Col_1)
```

```python
# Slicing, head and tail
# with head, not all values are printed, only first five
# with tail, last 5 lines
print(df.head())
```

```python
print(df.tail())
```

```python
print(df.tail(2))
```

```python
# indexes
print(df.index)
```

```python

# visualization
import datetime
import matplotlib.pyplot as plt
from pandas_datareader import data, wb
from matplotlib import style

style.use('fivethirtyeight')

start = datetime.datetime(2010, 1, 1)
end = datetime.datetime(2015, 7, 28)

# leveraging existing apis from the web to use as a dataframe, here source is yahoo
att = data.DataReader("T", "yahoo", start, end)

print(att.head())
```

```python
att['Adj Close'].plot()    # represent as graph
plt.legend()
plt.show()
```

```python
att['High'].plot()
att['Low'].plot()
plt.legend()
plt.show()
```

```python
# another way to do this
att[['High', 'Low']].plot()
plt.legend()
plt.show()
```
