```python
import pandas as pd
import urllib.request
read = urllib.request.urlopen("https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1556037771&period2=1587660171&interval=1d&events=history")
df = pd.read_csv(read)
print(df.head())
df.to_pickle('df.pickle')
```

```python
# so what we did here is - used urllib to download a csv file and read the data in df.. also changed it as pickle file. Pickle file present in current path
```

```python
df.head()
```

```python
# Sorting Data (date wise)
df.sort_values('Date', ascending=False, inplace=True)
df.head()

# you will see change in date values that are displayed
```

```python
df.set_index('Date', inplace=True)
df.head()
```

```python
# lets reduce the amount of data
df = df['Adj Close']
df.head()
```

```python
import matplotlib.pyplot as plt
df.plot()
plt.show()
```

```python
# multiple sorting
school = {
    'Name':['Je', 'Ak', 'Ank', 'Neh', 'Nbg', 'Zack', 'Sheldon', 'Sass', 'Tran', 'Jack'],
    'Age':[19, 19, 17, 18, 16, 14, 17, 20, 19, 15],
    'Grade':[9,9,10,12,8,8,7,12,10, 9]
}
df = pd.DataFrame(school)
df
```

```python
df2 = df.sort_values('Age')
df2
```

```python
df2 = df.sort_values(['Grade', 'Age'])
df2
```

```python
df2 = df.sort_values(['Grade', 'Age', 'Name'], ascending=False)
df2
```

```python
df2 = df.sort_values(['Grade', 'Age', 'Name'], ascending=[False, True, True])
df2
```
