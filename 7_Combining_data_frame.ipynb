```python
# combining data frames
# concatenation
import pandas as pd
df1 = pd.DataFrame({
    'Temp': [56,51,50,58],
    'Humidity': [30,40,45,48],
    'Precip': [0,0,0,1]
}, index = [0,1,2,3])

df2 = pd.DataFrame({
    'Temp': [76,77,78,79],
    'Humidity': [31,32,34,45],
    'Precip': [1,1,1,0]
}, index = [4,5,6,7])

df3 = pd.DataFrame({
    'Temp': [61,65,68,67],
    'Humidity': [43,44,47,40],
    'Precip': [0,0,1,1]
}, index = [8,9,10,11])
```

```python
concat = pd.concat([df1, df2, df3])
concat
```

```python
df4 = pd.DataFrame({
    'Temp': [61,65,68,67],
    'Humidity': [43,44,47,40],
    'Wind': [10,10,11,21]
}, index = [8,9,10,11])
```

```python
concat = pd.concat ([df1,df2,df4])
concat
```

```python
# now we will talk about merging data frames.
import pandas as pd
df1 = pd.DataFrame({
    'Temp': [56,51,50,58],
    'Humidity': [30,40,45,48],
    'Precip': [0,0,0,1]
})

df2 = pd.DataFrame({
    'Temp': [56,51,50,58],
    'Humidity': [31,32,34,45],
    'Precip': [1,1,1,0]
})

print(pd.merge(df1, df2, on='Temp'))
```

```python
import pandas as pd
df1 = pd.DataFrame({
    'Temp': [56,51,50,58],
    'Humidity': [30,40,45,48],
    'Precip': [0,0,0,1]
})

df2 = pd.DataFrame({
    'Temp': [56,51,50,58],
    'Cloudy': [31,32,34,45],
    'Wind': [1,1,1,0]
})

print(pd.merge(df1, df2, on='Temp'))
```
